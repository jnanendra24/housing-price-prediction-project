the above features have a very large number of missing values. So we drop them

```python
# dropping the columns

hsdata = hsdata.drop(['Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature'],axis = 1)
hsdata.info()
```

```python
# find % null values of remaining variables

percent_null2 = round(hsdata.isnull().sum()/len(hsdata) *100,2)
percent_null2 = pd.DataFrame(percent_null2,columns=['% null values'])
rem_null = percent_null2[percent_null2['% null values']>0]
rem_null
```

```python
# identifying categorical and numerical variables

# identifying categorical with null values
cat_vars = ['MasVnrType', 'BsmtCond', 'BsmtQual', 'BsmtExposure', 'GarageCond', 'GarageType', 'GarageQual', 'GarageFinish', 'BsmtFinType1', 'Electrical', 'BsmtFinType2']

# identifying numerical with null values

num_vars = ['LotFrontage','MasVnrArea','GarageYrBlt']

```

filling categorical values with mode
    

```python
for i in cat_vars:
    hsdata[i] = hsdata[i].fillna(hsdata[i].mode()[0])
```

filling continuous variables using KNN imputer

```python
from sklearn.impute import KNNImputer

imputer_knn = KNNImputer(missing_values=np.nan)

# Fill the missing values for cont_vars
for i in num_vars:
    hsdata[i] = imputer_knn.fit_transform(hsdata[[i]])
    hsdata[i] = hsdata[i].astype (int)
hsdata.sample(10)
```

```python
# confirming that there are no null values entirely

hsdata.isnull().sum()
```

Relationship between variables so that we can drop unnecessary ones

```python
hsdata.corrwith(hsdata['SalePrice'])
```

LowQualFinSF,MiscVal,BsmtFinSF2,PoolArea,3SsnPorch,ScreenPorch have very low correlation with the target variable so dropping them.

```python
# dropping the columns
hsdata = hsdata.drop(['LowQualFinSF','MiscVal','BsmtFinSF2','PoolArea','3SsnPorch','ScreenPorch'],axis=1)
```

```python
#step-7
#Outliers checking
```

```python
# plotting histogram for the target variable.

sns.histplot(hsdata['SalePrice'],kde=True)
```

```python
# boxplot for checking outliers

sns.boxplot(hsdata['SalePrice'])
```

```python
# getting outliers

first_quantile=hsdata["SalePrice"].quantile(.25)
third_quantile=hsdata["SalePrice"].quantile(.75)
IQR=third_quantile - first_quantile

#getting upper and lower bounds

Upper_bound=round(third_quantile + 1.5 * IQR ,3) 
Lower_bound=round(third_quantile - 1.5 * IQR ,3) 
```

```python
#Eliminating the outliers

hsdata = hsdata[(hsdata.SalePrice > Lower_bound) & (hsdata.SalePrice < Upper_bound)]
```

```python
sns.histplot(hsdata['SalePrice'],kde=True)
```

```python
#step-8,9
# Datatype conversion and encoding techinque
```

```python
cat_vars = hsdata.columns[hsdata.dtypes == 'object']
num_vars = hsdata.columns[hsdata.dtypes != 'object'][:-1]
```

```python
hsdata.describe(include="object")
```

```python
# encoding the categorical variable One hot encoding

hsdata = pd.get_dummies(hsdata,columns=cat_vars)
```

```python
hsdata.info()
```

# defining independent and Target Variables

```python
#step-10
IndepVar = []
for col in hsdata.columns:
    if col != 'SalePrice':
        IndepVar.append(col)
TargetVar = 'SalePrice'

x = hsdata[IndepVar]
y = hsdata[TargetVar]
```

# spliting the data into train and test data

```python
#step-11
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

# Scaling technique for continuous variables using StandardScaler

```python
#step-12
from sklearn.preprocessing import StandardScaler
mmscaler = StandardScaler()
x_train[num_vars] = mmscaler.fit_transform(x_train[num_vars])
x_train = pd.DataFrame(x_train)

x_test[num_vars] = mmscaler.fit_transform(x_test[num_vars])
x_test = pd.DataFrame(x_test)
```

# Model Building


KNN Algorithm

```python
#step-13
KNN_Results = pd.read_csv(r"KNNResults.csv",header=0)
```

```python

# Bild KNN Model

from sklearn.neighbors import KNeighborsRegressor

from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_absolute_percentage_error,r2_score

import sklearn.metrics as metrics

for a in range(1, 21, 1):
    
    k = a
    
    # Build the model
    
    ModelKNN = KNeighborsRegressor(n_neighbors=k)
    
    # Train the model
    
    ModelKNN.fit(x_train, y_train)
    
    # Predict the model
    
    y_pred = ModelKNN.predict(x_test)
    print('KNN_K_value = ', a)
    
    # Print the model name
    
    print('Model Name: ', ModelKNN)
    MAE = round(metrics.mean_absolute_error(y_test, y_pred), 3)
    MSE = round(metrics.mean_squared_error(y_test, y_pred),3)
    RMSE = round (np.sqrt(metrics.mean_squared_error(y_test, y_pred)),3)
    R2_score = round(metrics.r2_score (y_test, y_pred),6)
    MAPE = round(metrics.mean_absolute_percentage_error(y_test, y_pred)*100,3)
    RMSLE = round(np.log(np.sqrt(metrics.mean_squared_error(y_test, y_pred))),3)
    Adj_R_Square = round(1 - (1-R2_score)*(len(y)-1)/(len(y)-x.shape[1]-1),6)
    print('Mean Absolute Error (MAE):', MAE)
    print('Mean Squared Error (MSE):',MSE)
    print('Root Mean Squared Error (RMSE):', RMSE)
    print('R2_score:',R2_score)
    print('Mean Absolute Percentage Error (MAPE):',MAPE , '%')
    print('Root Mean Squared Log Error (RMSLE):',RMSLE)
    print('Adj R Square: ', Adj_R_Square)
    #------------------------------------------------------------------------------
    new_row = {'Model Name' : ModelKNN,
               'KNN K Value' : a,
               'Mean_Absolute_Error_MAE' : MAE,
               'Adj_R_Square' : Adj_R_Square,
               'Root_Mean_Squared_Error_RMSE' : RMSE,
               'Mean_Absolute_Percentage_Error_MAPE' : MAPE,
               'Mean_Squared_Error_MSE' : MSE,
               'Root_Mean_Squared_Log_Error_RMSLE' : RMSLE,
               'R2_score' : R2_score,
               }
    KNN_Results = KNN_Results.append(new_row, ignore_index=True)
    #------KNN_Results------------------------------------------------------------------------
```

```python
# Training the SVM algorithm

from sklearn.svm import SVR

ModelSVM1 = SVR(kernel='linear', C=1.0, epsilon=0.1)
ModelSVMGaussian = SVR(kernel='rbf', C=1.0, epsilon=0.1, gamma='scale')
ModelSVMPoly = SVR(kernel='poly', degree=2, C=1.0, epsilon=0.1)
ModelSVMSig = SVR(kernel='sigmoid', C=1.0, epsilon=0.1, gamma='scale', coef0=0.0)

# Train the model
ModelSVM1.fit(x_train,y_train)
ModelSVMGaussian.fit(x_train, y_train)
ModelSVMPoly.fit(x_train, y_train)
ModelSVMSig.fit(x_train, y_train)

# Predict the model with test data set
SVMS = [ModelSVM1,ModelSVMGaussian,ModelSVMPoly,ModelSVMSig]
for model in SVMS:
    y_pred = model.predict(x_test)
    # Print the model name
    print('Model Name: ', model)
    MAE = round(metrics.mean_absolute_error(y_test, y_pred), 3)
    MSE = round(metrics.mean_squared_error(y_test, y_pred),3)
    RMSE = round (np.sqrt(metrics.mean_squared_error(y_test, y_pred)),3)
    R2_score = round(metrics.r2_score (y_test, y_pred),6)
    MAPE = round(metrics.mean_absolute_percentage_error(y_test, y_pred)*100,3)
    RMSLE = round(np.log(np.sqrt(metrics.mean_squared_error(y_test, y_pred))),3)
    Adj_R_Square = round(1 - (1-R2_score)*(len(y)-1)/(len(y)-x.shape[1]-1),6)
    print('Mean Absolute Error (MAE):', MAE)
    print('Mean Squared Error (MSE):',MSE)
    print('Root Mean Squared Error (RMSE):', RMSE)
    print('R2_score:',R2_score)
    print('Mean Absolute Percentage Error (MAPE):',MAPE , '%')
    print('Root Mean Squared Log Error (RMSLE):',RMSLE)
    print('Adj R Square: ', Adj_R_Square)
    #------------------------------------------------------------------------------
    new_row = {'Model Name' : model,
               'Mean_Absolute_Error_MAE' : MAE,
               'Adj_R_Square' : Adj_R_Square,
               'Root_Mean_Squared_Error_RMSE' : RMSE,
               'Mean_Absolute_Percentage_Error_MAPE' : MAPE,
               'Mean_Squared_Error_MSE' : MSE,
               'Root_Mean_Squared_Log_Error_RMSLE' : RMSLE,
               'R2_score' : R2_score,
               }
    SVMResults = SVMResults.append(new_row, ignore_index=True)
```

# Comparing different algorithms

```python
RGRResults = pd.read_csv(r"RGRResults.csv")
```

```python
# Build the Regression / Regressor models

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import BayesianRidge
from sklearn.svm import SVR

# Create objects of Regression / Regressor models with default hyper-parameters

ModelMLR = LinearRegression()
ModelDCR = DecisionTreeRegressor()
ModelRFR = RandomForestRegressor()
ModelETR = ExtraTreesRegressor()
ModelKNN = KNeighborsRegressor(n_neighbors=5)
ModelBRR = BayesianRidge()
ModelSVR = SVR(kernel='linear', C=1.0, epsilon=0.1, gamma='scale')

# Evalution matrix for all the algorithms

MM = [ModelMLR, ModelDCR, ModelRFR, ModelETR, ModelKNN, ModelBRR, ModelSVR] 

for models in MM:
    
    # Fit the model with train data
    
    models.fit(x_train, y_train)
    
    # Predict the model with test data

    y_pred = models.predict(x_test)
    
    # Print the model name
    
    print('Model Name: ', models)
    
    # Evaluation metrics for Regression analysis
    MAE = round(metrics.mean_absolute_error(y_test, y_pred), 3)
    MSE = round(metrics.mean_squared_error(y_test, y_pred),3)
    RMSE = round (np.sqrt(metrics.mean_squared_error(y_test, y_pred)),3)
    R2_score = round(metrics.r2_score (y_test, y_pred),6)
    MAPE = round(metrics.mean_absolute_percentage_error(y_test, y_pred)*100,3)
    RMSLE = round(np.log(np.sqrt(metrics.mean_squared_error(y_test, y_pred))),3)
    Adj_R_Square = round(1 - (1-R2_score)*(len(y)-1)/(len(y)-x.shape[1]-1),6)
    print('Mean Absolute Error (MAE):', MAE)
    print('Mean Squared Error (MSE):',MSE)
    print('Root Mean Squared Error (RMSE):', RMSE)
    print('R2_score:',R2_score)
    print('Mean Absolute Percentage Error (MAPE):',MAPE , '%')
    print('Root Mean Squared Log Error (RMSLE):',RMSLE)
    print('Adj R Square: ', Adj_R_Square)
    
    #-------------------------------------------------------------------------------------------
    new_row = {'Model Name' : models,
               'Mean_Absolute_Error_MAE' : metrics.mean_absolute_error(y_test, y_pred),
               'Adj_R_Square' : round(Adj_R_Square,4),
               'Root_Mean_Squared_Error_RMSE' : RMSE,
               'Mean_Absolute_Percentage_Error_MAPE' : MAPE,
               'Mean_Squared_Error_MSE' : MSE,
               'Root_Mean_Squared_Log_Error_RMSLE': RMSLE,
               'R2_score' : round(R2_score,4)}
    RGRResults = RGRResults.append(new_row, ignore_index=True)
    #-----------------------------------------------------------------------------------------



```
