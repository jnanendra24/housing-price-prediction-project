```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import pandasql as psql
import warnings

warnings.filterwarnings("ignore")
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",500)
```

```python
hsdata_train = pd.read_csv(r"train.csv",header=0)
hsdata_test = pd.read_csv(r"test.csv",header=0)
hsdata_test_results = pd.read_csv(r"sample_submission.csv",header=0)
hsdata_test = pd.merge(hsdata_test,hsdata_test_results,on="Id")
hsdata = pd.concat([hsdata_train,hsdata_test],axis=0).reset_index(drop=True)
hsdata_bk = hsdata.copy()
hsdata.head()
```

```python
# deleting id which doesn't effect Target Variable SalePrice

hsdata = hsdata.drop(["Id"],axis=1)
```

```python
# Checking if there are any null values

hsdata.isnull().sum()
```

```python
# Calculating the percentage of null values

percent_null = round(hsdata.isnull().sum()/len(hsdata) *100,2)
percent_null = pd.DataFrame(percent_null,columns=['% null values'])
percent_null[percent_null['% null values']>0]
```

```python
# display the variables with more than 45 % of null values

percent_null[percent_null['% null values'] > 45]
```

the above features have a very large number of missing values. So we drop them

```python
hsdata = hsdata.drop(['Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature'],axis = 1)
hsdata.info()
```

```python
# find % null values of remaining variables

percent_null2 = round(hsdata.isnull().sum()/len(hsdata) *100,2)
percent_null2 = pd.DataFrame(percent_null2,columns=['% null values'])
rem_null = percent_null2[percent_null2['% null values']>0]
rem_null
```

```python
rem_null.index
```

still there are 29 columns with missing values

```python
# identifying categorical and numerical variables
cat_vars = hsdata.columns[hsdata.dtypes == 'object']
num_vars = hsdata.columns[hsdata.dtypes != 'object']

# identifying categorical with null values
cat_vars = set(rem_null.index).intersection(cat_vars)

# identifying numerical with null values

num_vars = set(rem_null.index).intersection(num_vars)

```

filling categorical values with mode
    

```python
for i in cat_vars:
    hsdata[i] = hsdata[i].fillna(hsdata[i].mode()[0])
```

```python
hsdata[cat_vars].isnull().sum()
```

filling numerical variables with mean

```python
for i in num_vars:
    hsdata[i] = hsdata[i].fillna(hsdata[i].mean())
```

```python
hsdata[num_vars].isnull().sum()
```

Outliers checking


```python
# Gaussian distribution (also known as normal distribution) is a bell-shaped curve

from scipy.stats import norm

plt.hist(hsdata.SalePrice, bins=20, rwidth=0.8, density=True)
plt.xlabel('SalePrice')
plt.ylabel('Count')

rng = np.arange(hsdata.SalePrice.min(), hsdata.SalePrice.max(), 0.1)
plt.plot(rng, norm.pdf(rng, hsdata.SalePrice.mean(),hsdata.SalePrice.std()))
```

```python
sns.histplot(hsdata['SalePrice'],kde=True)
```

```python
from scipy import stats
plt.figure(figsize=(13,5))
plt.subplot(1,3,1)
sns.distplot(hsdata["SalePrice"],kde=False, fit=stats.johnsonsu,color="red")
plt.title("Johnson Transformation",fontweight="black",size=20,pad=20)

plt.subplot(1,3,2)
sns.distplot(hsdata["SalePrice"],kde=False, fit=stats.norm,color="green")
plt.title("Normal Transformation",fontweight="black",size=20,pad=20)

plt.subplot(1,3,3)
sns.distplot(hsdata["SalePrice"],kde=False,fit=stats.lognorm,color="blue")
plt.title("Log Normal Transformation",fontweight="black",size=18,pad=20)
plt.tight_layout()
plt.show()
```

```python
# getting outliers
first_quantile=hsdata["SalePrice"].quantile(.25)
third_quantile=hsdata["SalePrice"].quantile(.75)
IQR=third_quantile - first_quantile

Upper_bound=round(third_quantile + 1.5 * IQR ,3) 
Lower_bound=round(third_quantile - 1.5 * IQR ,3) 
hsdata[(hsdata.SalePrice < Lower_bound) | (hsdata.SalePrice > Upper_bound)]
```
